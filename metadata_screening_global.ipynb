```python
import pandas as pd

folder_path = 'data/global/'
file = folder_path + 'df_evaluated_global.csv'
```

```python
df_global = pd.read_csv(file)
df_global
```

```python
df_global = df_global[df_global["Include"] != "no"].reset_index(drop=True)
df_global
```

```python
df_final = df_global[df_global["Data Source"].isna()].reset_index(drop=True)
df_final
```

```python
unique_sources = df_global['Data Source'].dropna().unique().tolist()
unique_sources = [source.strip() for item in unique_sources for source in item.split(';')]
unique_sources = list(set(unique_sources))

print(f"The total number of sources is: {len(unique_sources)}\n")
unique_sources
```

```python
unique_sources.sort(key=len)
filtered_sources = []

for source in unique_sources:
    if not any(source.startswith(existing) for existing in filtered_sources):
        filtered_sources.append(source)

print(f"The total number of filtered sources is: {len(filtered_sources)}\n")
filtered_sources
```
