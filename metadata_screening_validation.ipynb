```python
import pandas as pd

folder_path = 'data/validation/'
file = folder_path + 'df_evaluated_validation.csv'
```

```python
df_validation = pd.read_csv(file)
df_validation
```

```python
df_validation = df_validation[df_validation["Include"] != "no"].reset_index(drop=True)
df_validation
```

```python
df_final = df_validation[df_validation["Data Source"].isna()].reset_index(drop=True)
df_final
```

```python
unique_sources = df_validation['Data Source'].dropna().unique().tolist()
unique_sources = [source.strip() for item in unique_sources for source in item.split(';')]
unique_sources = list(set(unique_sources))

print(f"The total number of sources is: {len(unique_sources)}\n")
unique_sources
```

```python
unique_sources.sort(key=len)
filtered_sources = []

for source in unique_sources:
    if not any(source.startswith(existing) for existing in filtered_sources):
        if (source.startswith("MIT") and source != "MIT Olympic Scoring") \
           or (source.startswith("UN") and source != "UNLV Olympic Scoring"):
            continue
        filtered_sources.append(source)

print(f"The total number of filtered sources is: {len(filtered_sources)}\n")
filtered_sources
```

```python
original_list = [
    'MIT Olympic Scoring',
    'UNLV Olympic Scoring',
    'JIGSAWS',
    'UMONS-TAICHI',
    'Walking Gait',
    'UI-PRMD',
    'AQA-7',
    'MTL-AQA',
    'FIS-V',
    'Rhythmic Gymnastics',
    'TASD-2',
    'PISA',
    'FineDiving',
    'NETS',
    'PaSk',
    'FineFS',
    'MMASD',
    'LOGO'
]

print(f"The total number of original sources is: {len(original_list)}\n")
```

```python
set_filtered = set(filtered_sources)
set_original = set(original_list)

# Elements in common
common_elements = set_filtered & set_original

# Elements unique to each list
only_in_filtered = set_filtered - set_original
only_in_original = set_original - set_filtered


print("Elements in common:", common_elements)
print("Number of elements in common:", len(common_elements))

print("\nElements in filtered_sources but not in original_list:", only_in_filtered)
print("Number:", len(only_in_filtered))

print("\nElements in original_list but not in filtered_sources:", only_in_original)
print("Number:", len(only_in_original))
```
